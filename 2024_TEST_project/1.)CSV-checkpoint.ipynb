# Motorsiklet veri seti için görselleştirme

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sütun isimlerinden olası boşlukları ve hatalı karakterleri kaldır
df.columns = df.columns.str.strip()


plt.figure(figsize=(10, 6))
sns.barplot(x='Marka', y='fiyat', data=df)
plt.title('Markalara Göre Fiyat Dağılımı')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='yil', y='fiyat', data=df)
plt.title('Yıllara Göre Fiyat Dağılımı')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='model', y='fiyat', hue='Marka', data=df, s=100)
plt.title('Marka ve Model Bazında Fiyat Karşılaştırması')
plt.xticks(rotation=45)
plt.show()
```

# Müşteri veri seti ödev

```python
veri = pd.read_csv('Employee.csv')
```

```python
veri = pd.DataFrame(veri)
veri
```

```python
#cinsiyete göre yas dağilimi

import matplotlib.pyplot as plt
import seaborn as sns

# Seaborn ile cinsiyete göre yaş dağılımı
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='Age', data=veri)
plt.title('Cinsiyete Göre Yaş Dağılımı')
plt.show()
```

```python
#şehre göre çalışan dagilimi

plt.figure(figsize=(8, 6))
sns.countplot(x='City', data=veri)
plt.title('Şehre Göre Çalışan Sayıları')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Education', hue='PaymentTier', data=veri)
plt.title('Eğitim Seviyesine Göre Ödeme Katmanları')
plt.show()
```
