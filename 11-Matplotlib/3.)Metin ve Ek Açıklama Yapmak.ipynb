# Metin ve Ek Açıklama Yapmak

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
fig, ax = plt.subplots(facecolor='lightblue')

ax.axis([0, 10, 0, 10]) # x-baslangıç,x-bitis / y-baslangıç,y-bitis

ax.text(2, 6, "A")
ax.text(4, 2, "(4, 2)")
ax.text(8, 2, "(8, 2)")

ax.grid(True)
```

```python

fig, ax = plt.subplots()

x = np.linspace(5, 15, 1000)

ax.plot(x, x)

ax.axis([0, 20, 0, 20])


ax.annotate('minimum', xy=(5, 5), xytext=(10, 5),
            arrowprops=dict(facecolor='red'))



ax.annotate('maximum', xy=(15, 15), xytext=(7,10),
            arrowprops=dict(facecolor='red',
                            arrowstyle="->",
                            connectionstyle="angle3,angleA=90,angleB=0"))


ax.annotate('Orjin',xy=(0, 0), xytext=(30,50), 
            textcoords='offset points',
            size=12, 
            bbox=dict(boxstyle="round", fc='lightgreen'),
            arrowprops=dict(arrowstyle="wedge",
                            fc='lightgreen'))

ax.grid(True)
```
