# Pie

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python

labels = 'Keçiler', 'Kuzular', 'İnekler', 'Köpekler'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # sadece 2. dilimi patlattık

fig1, ax1 = plt.subplots()

ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90) # startangle ile döndürdük
ax1.axis('equal')  # Eşit en boy oranı, turtanın daire şeklinde çizilmesini sağlar.

plt.show()
```
