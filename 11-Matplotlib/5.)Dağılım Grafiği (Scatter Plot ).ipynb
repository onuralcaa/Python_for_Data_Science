# Dağılım Grafiği "Scatter Plot "

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
np.random.seed(1) # tohum değerini biz belirledik, rastgele oluşan sayılar zinciri hepimizde aynı çıksın diye


N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = 300 * np.random.rand(N) 
```

```python
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```
